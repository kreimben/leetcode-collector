# Refine for CodeMind LLM datasets.

### In this file, I will merge 4 csv files into huggingface parquet file.

```python
import pandas as pd
from datasets import load_dataset
```

```python
caption = load_dataset("kreimben/leetcode_with_youtube_captions")['train']
caption = pd.DataFrame(caption)

print(f'{len(caption)=}')

# remove redundant columns
caption.drop(columns=['video_id', 'datetime', 'title'], inplace=True)
caption.rename(columns={'leetcode_number': 'id', 'content': 'cc_content'}, inplace=True)

caption.sample(5)
# Thumbnail is just for judge whether it's wrong video!
# And have to remove when training!
```

```python
# which is actually from https://www.kaggle.com/datasets/manthansolanki/leetcode-questions
leetcode_info = pd.read_csv('leetcode_questions.csv')

print(f'{len(leetcode_info)=}')

leetcode_info.drop(columns=['Question Slug', 'Similar Questions Text'], inplace=True)
leetcode_info.rename(
    columns={
        'Question ID': 'id', 
        'Question Title': 'title',
        'Question Text': 'question_content',
        'Topic Tagged text': 'tag',
        'Difficulty Level': 'level',
        'Success Rate': 'success_rate',
        'total submission': 'total_submission',
        'total accepted': 'total_accepted',
        'Likes': 'question_likes',
        'Dislikes': 'question_dislikes',
        'Hints': 'question_hints',
        'Similar Questions ID': 'similar_question_ids'
    }, 
    inplace=True
)

leetcode_info.head(5)
```

```python
question_and_solution = load_dataset("greengerong/leetcode")['train']
question_and_solution = pd.DataFrame(question_and_solution)

print(f'{len(question_and_solution)=}')

question_and_solution.drop(columns=['slug', 'difficulty'], inplace=True)
question_and_solution.rename(columns={'content': 'question_content'}, inplace=True)

question_and_solution.head(5)
```

```python
a = leetcode_info.loc[100, 'question_content']
b = question_and_solution.loc[100, 'question_content']

a, b
```

### After comparing with `leetcode_info` and `question_and_solution`, I decided to use `question_and_solution` for reason rich information.

```python
leetcode_info.drop(columns=['question_content'], inplace=True)
leetcode_info.sample(5)
```

```python
# merge like SQL's join.

result = pd.merge(question_and_solution, leetcode_info, on='id', how='left')
result.sort_values(by='id', inplace=True)
result.sample(5) 
# the index and id don't match because some of questions are not collected.
```
